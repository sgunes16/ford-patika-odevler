6-Her kıta için şarap tüketiminin betimsel istatiğini yapınız.

```python
drinks_df[["country","wine_servings"]].describe()
```

7-Her kıta için ve her alkol cinsi için ortalama alkol tüketimini gösteriniz. 

```python
drinks_df.groupby("continent").mean()
```

8-Her kıta için ve aşağıdaki alkol cinsleri için belirtilen hesaplamaları gerçekleştiriniz.

beer_servings - std
spirit_servings - median 
wine_servings - min, mean, max
total_litres_of_pure_alcohol - mean, count

```python
drinks_df.groupby("continent").beer_servings.std()
drinks_df.groupby("continent").spirit_servings.median()
drinks_df.groupby("continent").wine_servings.min()
drinks_df.groupby("continent").wine_servings.mean()
drinks_df.groupby("continent").wine_servings.max()
drinks_df.groupby("continent").total_litres_of_pure_alcohol.mean()
drinks_df.groupby("continent").total_litres_of_pure_alcohol.count()
```

9- Her alkol cinsinin en çok tüketildiği ülkeyi bulun. 

```python
drinks_df.sort_values(by='beer_servings', ascending=False).head(1)['country']
drinks_df.sort_values(by='wine_servings', ascending=False).head(1)['country']
drinks_df.sort_values(by='spirit_servings', ascending=False).head(1)['country']
```

10- Yeni değişkenler üreterek her alkol cincisin ilgili ülkedeki toplam alkol tüketimi içerisindeki oranlarını yazdırınız.  

```python
drinks_df['beer_proportion'] = drinks_df['beer_servings']/(drinks_df['beer_servings']+drinks_df['spirit_servings']+drinks_df['wine_servings'])
drinks_df['wine_proportion'] = drinks_df['wine_servings']/(drinks_df['beer_servings']+drinks_df['spirit_servings']+drinks_df['wine_servings'])
drinks_df['spirit_proportion'] = drinks_df['spirit_servings']/(drinks_df['beer_servings']+drinks_df['spirit_servings']+drinks_df['wine_servings'])

```
